```python
# i din't 
df= pd.read_csv('train.csv')
pd.set_option('display.max_columns',None)
df.head()
```

```python
feature_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1]
pd.set_option('display.max_columns',None)
for feature in feature_with_nan:
    print(feature,':',np.round(df[feature].isnull().mean(),4),'%missing value')
```

```python
for feature in feature_with_nan:
    data = df.copy()
    
    data[feature]= np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numcal_fer = [feature for feature in df.columns if df[feature].dtype!='O']
df[numcal_fer].head()
```

```python
#tempoyer varable like datae and year
yrear_feature =[feature for feature in numcal_fer if 'Yr' in feature or 'Year'in feature]
yrear_feature
```

```python
for feature in yrear_feature:
    print(feature,df[feature].unique())
```

```python
for feature in yrear_feature:
    if feature!='YrSold':
        data = df.copy()
        data[feature]=data['YrSold']-data[feature]


        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrat_fe=[feature for feature in numcal_fer if len(df[feature].unique())<25 and feature not in yrear_feature+['Id']]
print('discreat varablr count:{}'.format(len(discrat_fe)))
```

```python
df[discrat_fe].head()
```

```python
for feature in discrat_fe:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
countunes_far=[feature for feature in numcal_fer if feature and feature not in discrat_fe+ yrear_feature+['Id']]
df[countunes_far].head()
```
