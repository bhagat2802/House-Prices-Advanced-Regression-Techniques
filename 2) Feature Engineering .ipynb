```python
df = pd.read_csv('train.csv')
df.head()
```

```python
#miising of catogrical fea

feature_nan_cat =[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
for feature in feature_nan_cat:
    print('{}: {}% missing values'.format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
 #replacing missing features 
def replace_cat_fer(df,feature_nan_cat):
    data=df.copy()
    data[feature_nan_cat]=df[feature_nan_cat].fillna('missing')
    return data
df= replace_cat_fer(df,feature_nan_cat)
    
```

```python
df[feature_nan_cat].isnull().sum()
```

```python
df.head()
```

```python
#numical feture
feature_nan_num =[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']
for feature in feature_nan_num:
    print('{}: {}% missing values'.format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#numacal missing valu handel
for feature in feature_nan_num:
    median_val=df[feature].median()
    
    df[feature +'fea'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_val,inplace=True)
    
    
     
```

```python
df.head(50)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df[feature]=df['YrSold']-df[feature]
    
```

```python
df.head()
```

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

# Handling Rare Categorical Feature


```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']

```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
    
```

```python
df.head()
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head(50)
```

# feacture scaling 

```python
feacuture_scale =[feature for feature in df.columns if feature not in ['Id','SalePrice']  ]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feacuture_scale])
```

```python
scaler.transform(df[feacuture_scale])
```
